```python
import pandas as pd
import numpy as np
import plotly_express as px
```

```python
# Suicide Data
games = pd.read_csv("vgsales-12-4-2019.csv")
games.head()
```

```python
games_by_genre = games.groupby(['Genre','ESRB_Rating','NA_Sales']).aggregate(np.sum).reset_index()
games_by_genre.head()
```

```python
releases_per_year_2007 = games_by_genre[games_by_genre['Year']>1998]

px.scatter(releases_per_year_2007,x = 'Genre', y = 'NA_Sales')
```

```python
px.scatter(releases_per_year_2007,x = 'Genre', y = 'NA_Sales',color='ESRB_Rating')
```

```python
px.scatter(releases_per_year_2007,x = 'Genre', y = 'NA_Sales',
           color='ESRB_Rating',size ='NA_Sales')
```

```python
px.scatter(releases_per_year_2007,x = 'Genre', y = 'NA_Sales', size = 'NA_Sales', 
           color='ESRB_Rating',facet_col='ESRB_Rating')
```

```python
px.scatter(releases_per_year_2007,
           x = 'Genre', 
           y = 'NA_Sales',
           color='ESRB_Rating',
           size='NA_Sales',
           
           animation_frame='Year', 
           animation_group='ESRB_Rating',
           range_x = [0,10],
          range_y = [0,6],
           text='ESRB_Rating')
```
